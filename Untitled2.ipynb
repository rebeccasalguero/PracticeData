# Reorder the variables so that the key variable (part_id2) is first

### Before making column (part_id2) first in the dataset
### Store the column (part_id2) into a variable called key_var
### We can now drop the column (part_id2) from the entire dataset
### This is stored in variable df_drop

### Now we join the column (part_id2) that we stored as key_var to the dataset we stored in df_drop with the concat() function

```python
key_var = df['part_id2']

df_drop = df.drop(columns ='part_id2')

new_df = pd.concat([key_var, df_drop], axis = 1)
new_df
```

# Sort on the key variable (smallest to largest).
### In order to sort the data by a column, use the .sort_values() function 
### parameters like (by = ' ' ) within the function will allow you to specify what column 
### with a comma you can also add another rule/parameter
### in this case we can use ascending equal to true to return data in smallest to largest order

```python
sorted_data1 = new_df.sort_values(by = 'part_id2', ascending = True)
sorted_data1
```

# Recode the gender variable so that it is no longer a string variable
## Rewrite so that dataframe changes the gender column to 0s and 1s 
### F = 1
### M = 0

## From here on out, I will be refering to variable sorted_data2, which holds the dataset with the changes we made up to this point

```python
sorted_data2 = sorted_data1
sorted_data2['gender'] = sorted_data2['gender'].replace({'F': 1, 'M':0})
sorted_data2
```

# Finding the mean, st_dev, range of dataset(except random and insample) excluding missing values

### Before calcualting mean, standard dev, and range, we need to exclude the insample and random and store the Data in varibale calc
### calc is the variable that contains the columns/data we want
### We now can calculate the mean, standard dev and range of the data
### For organization, visual purposes, and readibility
### I stored the calculations as tables with the DataFrame() function, and relabled the columns accordingly with the rename() function
### I merged/joined all three tables(mean, st_dev, rang) using the concat() function 

```python
calc = pd.DataFrame(sorted_data2[sorted_data2.columns[:-2]])

mean = pd.DataFrame(calc.mean()).rename(columns = {0:'Mean'})
st_dev = pd.DataFrame(calc.std()).rename(columns = {0:'Standard Deviation'})
rang = pd.DataFrame(calc.max() - calc.min()).rename(columns = {0:'Range'})

frames = [mean, st_dev, rang]
table = pd.concat(frames, axis = 1)
table
```

# Use Gender as a grouping variable and find the respective mean, st dev, and range for each

### Before making calculations, I want to isolate the data based on gender
### I create two tables, one for female and one for male
### After tables are created I can start making the necessary calculations
### 

```python
female = calc[calc['gender'] == 1]
male = calc[calc['gender'] == 0]

fem_mean = female.mean()
fem_std = female.std()
fem_range = female.max() - female.min()

m_mean = male.mean()
m_std = male.std()
m_range = male.max() - male.min()
```

### Results will show the mean, st dev, and range for male and female

## For organization, visual purposes, and readibility
## I saved the calculations as a table(f_frame, m_frame) with the DataFrame() function and relabled the row in the table by using the index= parameter
## Added the title Female and Male to each

```python
f_frame = pd.DataFrame([fem_mean,fem_std, fem_range], index=['Mean', 'Standard Dev', 'Range'])
title_fem = f_frame.index.name = 'Female'
f_frame
```

```python
m_frame = pd.DataFrame([m_mean,m_std, m_range], index=['Mean', 'Standard Dev', 'Range'])
title_male = m_frame.index.name = 'Male'
m_frame
```
